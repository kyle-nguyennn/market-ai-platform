# Feature Store Demo

Shows how to compute features, write them to the online Redis cache, and retrieve them.

```python
import polars as pl  # noqa: F401

from libs.features.rolling import ret_1d, rolling_volatility  # noqa: F401
from libs.storage.redis_store import RedisStore

store = RedisStore()
print('Redis ping:', store.healthcheck())
```

```python
# Simulate writing features for ticker 'SPY'
features = {'ret_1d': 0.012, 'vol_20d': 0.18, 'mom_20d': 0.05, 'adv_20d': 85e6}
store.set_features('SPY', features, ttl_seconds=3600)
print('Written:', features)
```

```python
# Read back
cached = store.get_features('SPY')
print('Read back:', cached)
```
