# End-to-End Inference Demo

Walks through: load model artifact → construct feature vector → call inference-gateway → inspect response.

```python
import httpx

BASE_URL = 'http://localhost:8002'

# Check health
resp = httpx.get(f'{BASE_URL}/health')
print(resp.json())
```

```python
# Score a single entity
payload = {
    'model_name': 'xgb_alpha_v1',
    'entity_id': 'SPY',
    'features': {
        'ret_1d': 0.012,
        'vol_20d': 0.18,
        'mom_20d': 0.05,
        'adv_20d': 85000000.0,
    },
}
resp = httpx.post(f'{BASE_URL}/score/', json=payload)
print(resp.status_code, resp.json())
```
